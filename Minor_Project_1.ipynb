```python

top_list = []
for gen, num in Gen_Count[:50]:
    sub_df = df.loc[df["Genres"]==gen]    
    #count = len(sub_df)
    users = sum(sub_df["User Rating Count"])
    total = 0
    for rate, user in zip(sub_df["Average User Rating"], sub_df["User Rating Count"]):
        total += rate * user
    average_rating = round(total/users, 2)
    top_list.append([gen, num, average_rating, users])
```

```python
top_list
```

```python
top_df = pd.DataFrame(top_list)
top_df
```
