```python
import pandas as pd
music_data = pd.read_csv("music.csv")
music_data
```

```python
md = music_data
```

```python
mdin = md.drop(columns="gener")
mdin
```

```python
mdout = md.drop(columns=["AGE", "gender"])
mdout
```

```python
# we are implimenting "decision tree machine learning model"
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(mdin, mdout)
```

```python
md
```

```python
predictions = model.predict([[21, 1], [22, 0]])
predictions
```
