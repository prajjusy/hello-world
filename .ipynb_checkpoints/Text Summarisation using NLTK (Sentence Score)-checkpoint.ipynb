```python
from urllib import request
from bs4 import BeautifulSoup as bs
import re
import nltk
```

```python
url = "https://en.wikipedia.org/wiki/Machine_learning"
```

```python
htmlDoc = request.urlopen(url)
htmlDoc
```

```python
soup_object = bs(htmlDoc, "html.parser")
```

```python
paragraph_contents = soup_object.findAll("p")
```

```python
print(paragraph_contents)
```

```python
all_paragraph_content = ""
```

```python
for paragraph_content in paragraph_contents:
    all_paragraph_content += paragraph_content.text + "\n"
```

```python
print(all_paragraph_content)
```

```python
all_paragraph_content_cleaner_data = re.sub(r"\[[0-9]*\]", " ", all_paragraph_content)
all_paragraph_content_cleaned_data = re.sub(r"\s+", " ", all_paragraph_content_cleaner_data)
```

```python
all_paragraph_content_cleaned_data
```

```python
all_paragraph_content_cleaned_data = re.sub(r"\[^a-zA-Z]", " ", all_paragraph_content_cleaned_data)
all_paragraph_content_cleaned_data = re.sub(r"\s+", " ", all_paragraph_content_cleaned_data)
```

```python
all_paragraph_content_cleaned_data
```

```python
sentences_tokens = nltk.sent_tokenize(all_paragraph_content_cleaned_data)
```

```python
sentences_tokens
```

```python
words_tokens = nltk.word_tokenize(all_paragraph_content_cleaned_data)
words_tokens
```

```python
stop_words = nltk.corpus.stopwords.words("english")
```

```python
word_frequencies = {}
```

```python
for word in words_tokens:
    if word not in stop_words:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1
            
```

```python
word_frequencies
```

```python
maximum_frequency_word = max(word_frequencies.values())
maximum_frequency_word
```

```python
for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequency_word)
    
word_frequencies
```

```python
sentence_scores = {}

for sentence in sentences_tokens:
    for word in nltk.word_tokenize(sentence.lower()):
        if word in word_frequencies.keys():
            if (len(sentence.split(" ")) < 30):
                if sentence not in sentence_scores.keys():
                    sentence_scores[sentence] = word_frequencies[word]
                else:
                    sentence_scores[sentence] += word_frequencies[word]
                    
sentence_scores
```

```python
import heapq
```

```python
summary = heapq.nlargest(10, sentence_scores, key=sentence_scores.get)
summary
```

```python
sentence_scores['In weakly supervised learning, the training labels are noisy, limited, or imprecise; however, these labels are often cheaper to obtain, resulting in larger effective training sets.']
```
