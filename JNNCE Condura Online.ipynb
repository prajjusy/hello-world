```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import collections
```

```python
data = pd.read_excel("JNNCE Condura Online.xlsx")
```

```python
data
```

```python
activeness = collections.Counter(data["Active User Status"])
```

```python

sns.countplot(data["Active User Status"])
plt.plot()
print()
print()
print("\tTotal number of students:\t", activeness["Yes"]+activeness["No"])
print()
print("\tNumber of active studenst:\t", activeness["Yes"])
print("\tNumber of inactive studenst:\t", activeness["No"])
print()
```

```python
x = data[data["Active User Status"] == "Yes"]
x = x[x["Tasks Completed"]>0]

print("Number of students completed more than 100 tasks:\t", len(x[x["Tasks Completed"]>100]))
print("Number of students completed more than 150 tasks:\t", len(x[x["Tasks Completed"]>150]))
print("Number of students completed more than 200 tasks:\t", len(x[x["Tasks Completed"]>200]))
print("Number of students completed more than 250 tasks:\t", len(x[x["Tasks Completed"]>250]), "\n")
print("Number of students completed more than 500 tasks:\t", len(x[x["Tasks Completed"]>500]))


```
